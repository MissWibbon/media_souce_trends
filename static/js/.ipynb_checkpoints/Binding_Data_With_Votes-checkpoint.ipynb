```python
import json
import csv
```

```python
geo_json_file = "statesOutline.js"
```

```python
geo_json = json.load(open(geo_json_file, 'r'))
```

```python
voting_file = "../data/2008-2016-president.csv"
```

```python
years = ["2008", "2012", "2016"]
```

```python
state_votes = {}

with open(voting_file) as csv_file:
    csv_reader = csv.reader(csv_file)
    for row in csv_reader:
        if row[7] in ["republican", "democrat"]:
            # print(f"{row[0]}: {row[1]} {row[6]} {row[7]} {row[9]} ")
            state_name = row[1]
            if not state_name in state_votes:
                state_votes[state_name] = {}
            candidate = row[6]
            party = row[7]
            vote_count = row[9]
            year = row[0]

            if not year in state_votes[state_name]:
                print(f"adding {year} for {state_name}")
                state_votes[state_name][year] = {}
            else:
                print(f"skipping {year} for {state_name}")

            state_votes[state_name][year][candidate] = [party, vote_count]
```

```python
state_votes
```

```python
for state in geo_json['features']:
    state_name = state['properties']['NAME']
    if state_name in state_votes.keys():
        state['properties']['voting_data'] = state_votes[state_name]
```

```python
json.dump(geo_json,open('statesOutlineWithVotes.json', 'w+'))
```
