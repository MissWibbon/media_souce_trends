```python
import json
```

```python
geo_json_file = "statesOutline.js"
```

```python
geo_json = json.load(open(geo_json_file, 'r'))
```

```python
cnn_fox_file = "../data/CNNFOX.csv"
```

```python
# features year, data (cnn/fox)
```

```python
years = []
year = 2005
for i in range (0,15):
    year += 1 
    years.append(year)
```

```python
years
```

```python
state_elect = {}
with open(cnn_fox_file) as cnn_fox:
    for line in cnn_fox:
        if line.startswith('Region'):
            next
        else: 
            state_name = line.split(',')[0]
            state_elect[state_name] = {}
            percents = line.split(',', 1)[1]
            split_percents = percents.split(',')
            for year in years: 
                state_elect[state_name][year] = []
                print(split_percents)
                if len(split_percents) > 0:
                    for i in range(2):
                        state_elect[state_name][year].append(split_percents[i])
                    split_percents.pop(0)
                    split_percents.pop(0)
                    next
```

```python
state_elect
```

```python
for state in geo_json['features']:
    state_name = state['properties']['NAME']
    if state_name in state_elect.keys():
        state['properties']['election_data'] = state_elect[state_name]
```

```python
json.dump(geo_json,open('statesOutlineWithElection.json', 'w+'))
```
